```python
sns.scatterplot(data=df,x='sepal.length',y='sepal.width')
sns.scatterplot(data=df,x='petal.length',y='petal.width')
```

```python
import ipywidgets as widgets 

x1 = widgets.IntSlider(description="x-max", continuous_update=True)
y1 = widgets.IntSlider(description="y-max", continuous_update=True)
widgets.VBox([x1,y1])
```

```python
@widgets.interact(x1=7,y1=7)
def iris_scatter(x1,y1):
    f,ax = plt.subplots(1,1)
    ax.scatter(df['sepal.length'].values,df['sepal.width'].values,s=12)
    ax.scatter(df['petal.length'].values,df['petal.width'].values,s=12,color=(1,0,0))
    
    ax.set_xlim(-1,x1)
    ax.set_ylim(-1,y1)
```
